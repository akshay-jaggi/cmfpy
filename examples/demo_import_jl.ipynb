### Import Demo
**Import a model from cmf.jl and visualize the fit**

```python
import cmfpy
%matplotlib inline

data, hals_model = cmfpy.model.load_cmfjl_model("../../cmf.jl/hals_results.hd5")
_, mult_model = cmfpy.model.load_cmfjl_model("../../cmf.jl/mult_results.hd5")
_, annls_model = cmfpy.model.load_cmfjl_model("../../cmf.jl/annls_results.hd5")
```

```python
hals_model.plot(tmin=0, tmax=300);
```

```python
mult_model.plot(tmin=0, tmax=300);
```

```python
annls_model.plot(tmin=0, tmax=300);
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.imshow(data[hals_model.argsort_units(), :300])
plt.show()
```
